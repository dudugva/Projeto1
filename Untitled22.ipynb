```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


g = 9.8
#densidade do ar [kg/m^3]
p = 1.6
#massa do sistema, ou seja saltador+paraquedas [kg]
def mas(t):
    if t<30:
        massa=74 #aberto, llungo
    elif t>=30 and t<=64:
        massa=184 #brigando lungo
    else:
        massa=104
    return massa
#Cd será o coeficiente de arrasto que irá variar no tempo
#A por sua vez será a Área que também irá variar
def Cd(t):
    if t<64:
        coef = 1.3 #aberto, llungo
    else:
        coef = 0.42
    return coef

def Ay(t):
    if t < 19:
        Area = 1
    if t>=19 and t<30:
        Area = 0.05
    elif t>=30 and t<=40:
        Area=2
    else:
        Area=6.26*math.pi
    return Area

def Ax(t):
    if t < 19:
        Area = 0.05
    if t>=19 and t<30:
        Area = 0.05
    if t>=30 and t<=40:
        Area=0.1
    else:
        Area=8.31*math.pi
    return Area

def EqDif(y,t):
    X = y[0]
    Vx = y[1]
    Y = y[2]
    Vy = y[3]
    dxdt = Vx
    dVxdt = -(1/2)*((p*Cd(t)*Ax(t)*((Vx**2)+(Vy**2)))/mas(t))*(Vx/((math.sqrt((Vx**2)+(Vy**2)))))
    dydt = Vy
    dVydt = (-(1/2)*((p*Cd(t)*Ay(t)*((Vx**2)+(Vy**2)))/mas(t))*(Vy/((math.sqrt((Vx**2)+(Vy**2))))) - g)
    return [dxdt,dVxdt,dydt,dVydt]

X0 = 0
Vx0 = 200
Y0 = 4000
Vy0 = 0
C0 = [X0,Vx0,Y0,Vy0]
t=np.arange(0,120,0.01)
solução = odeint(EqDif,C0,t)
yy=[]
i=0
ti=[]
while i<11900:
    yy.append(solução[:,2][i])
    ti.append(i)
    i=i+1
    
plt.plot(t,solução[:,2],color="red")
plt.title("Trajetória do Saltador em Y por tempo")
plt.xlabel("Tempo [segundos]")
plt.ylabel("Altura [metros]")
plt.grid()
plt.show()

hrampa=(solução[:,2][9999])
```

```python
plt.plot(t,solução[:,3],color="blue")
plt.title("Velocidade por tempo durante a queda")
plt.xlabel("Tempo [segundos]")
plt.ylabel("Velocidade [m/s]")
plt.grid()
plt.show()

hrampa=(solução[:,2][9999])
```

```python
v0=(((solução[:,1][9999])**2)+((solução[:,3][9999])**2))**(1/2)
```

```python
t = np.arange(0,3.36,0.01)
#para 50m de rampa
r=31.846
omega0=v0/r
def Rampa(Y,t):

    teta = Y[0]
    omega = Y[1]
    
    dtetadt = omega
    domegadt = (-g*math.sin(teta))/r #aceleração tangencial
    return (dtetadt, domegadt)


S0=[math.radians(-45),omega0]
Solucao=odeint(Rampa,S0,t)
grau=[]
i=0
X=[]
Y=[]

while i<=335:
    grau.append(math.degrees(Solucao[:,0][i]))
    X.append(-r*math.sin(Solucao[:,0][0])+r*(math.sin(Solucao[:,0][i])))
    Y.append(hrampa+(r*math.cos(Solucao[:,0][0])-r*math.cos(Solucao[:,0][i])))
    yy.append(Y[i])
    
    if math.degrees(Solucao[:,0][i])>=45:
        omegafinal=Solucao[:,1][i]
        break
    i=i+1
plt.plot(X,Y)
plt.title("Trajetória sobre a rampa")
plt.xlabel("Tempo [segundos]")
plt.ylabel("Posição em y")
plt.show()
yy.append(Y)
vfinal=omegafinal*r
```

```python
def lanca(y,t):
    X = y[0]
    Vx = y[1]
    Y = y[2]
    Vy = y[3]
    dxdt = Vx
    dVxdt = 0 #(Vx/((math.sqrt((Vx**2)+(Vy**2)))))
    dydt = Vy
    dVydt =-g
    return [dxdt,dVxdt,dydt,dVydt]

X0 = 0
Vx0 = (vfinal/2)*(2**(1/2))
Y0 = hrampa
Vy0 = (vfinal/2)*(2**(1/2))
C0 = [X0,Vx0,Y0,Vy0]
t=np.arange(0,100,0.01)
solução = odeint(EqDif,C0,t)
i=0
x1=[]
y1=[]
while i<99000:
    x1.append(solução[:,0][i])
    y1.append(solução[:,2][i])
    yy.append(y1[i])
    if solução[:,2][i]<hrampa:
        print(i)
        break
    i=i+1
yy.append(y1)
ti=np.arange(0,124.16,0.1)
plt.plot(x1,y1,color="red")
plt.title("Trajetória de Bond após saltar da rampa")
plt.xlabel("Tempo [segundos]")
plt.ylabel("Altura - eixo y - [metros]")
plt.grid()
plt.show()
```
